```python
%python
from pyspark.sql.session import SparkSession
spark=SparkSession.builder.appName("SparkByExamples.com").getOrCreate()
df1=spark.read.csv("/Volumes/workspace/default/volumewe47_datalake/patients.csv",header="True")
display(df1)
                
```
