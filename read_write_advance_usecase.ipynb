####JSON Write Advanced Feature - Very Important, Important, Not Important (just try to know once for all)
**Very Important:** path, mode, compression
****Important:** **dateFormat, timestampFormat, lineSep, ignoreNullFields
**- Not Important:** encoding

```python
from pyspark.sql.functions import *
struct1="cid int,name string,amt double,dop date"
df1=spark.read.csv("/Volumes/workspace/default/volumewe47_datalake/malformeddata1.txt",header=False,sep=",",comment='#',quote="'",escape='~',schema=struct1,multiLine=True,ignoreLeadingWhiteSpace=True,ignoreTrailingWhiteSpace=True,dateFormat='yyyy-dd-MM')#permit malformed records and allow me to do a RCA (Root cause analysis)
#df2=df1.withColumn("dop",to_date(date_format(col("dop"),"yyyy-dd-MM"),"yyyy-MM-dd"))
df1.printSchema()
df1.show()
```

```python
df1.write.json("/Volumes/workspace/default/volumewe47_datalake/outputdata/jsonout",mode='overwrite',compression='gzip',ignoreNullFields='True',lineSep='~')
display(df1)
```

```python
df1.write.orc("/Volumes/workspace/default/volumewe47_datalake/outputdata/orcout",mode='overwrite',compression='snappy',partitionBy='dop')
display(df1)
df1.write.parquet("/Volumes/workspace/default/volumewe47_datalake/outputdata/parquetout",mode='overwrite',compression='snappy',partitionBy='dop')
display(df1)
display(df1.explain())
```
